```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
digits = load_digits()
digits.data.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
digits_std = StandardScaler().fit_transform(digits.data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_std, digits.target, stratify = digits.target, test_size=0.2, random_state=seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([Dense(100, input_dim=64, activation='relu'),
                    Dense(24, activation='relu'), Dense(1, activation='sigmoid')])
```

```python
model.compile('adam', 'binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
mc = ModelCheckpoint('digits_best.h5', monitor='val_loss', verbose=1, save_best_only=True)
es = EarlyStopping(patience = 30)
```

```python
hist = model.fit(X_train, y_train, validation_split=0.2, epochs=1000, batch_size=100, verbose=0, callbacks= [mc,es])
```

```python
model.evaluate(X_test, y_test)
```

```python
model.save('digits_best.h5')
```

```python
from tensorflow.keras.models import load_model
best_model = load_model('digits_best.h5')
best_model.evaluate(X_test, y_test)
```
