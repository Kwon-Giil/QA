```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
digits = load_digits()
digits.data.shape
```

 - 전처리

```python
# x데이터 표준화
from sklearn.preprocessing import StandardScaler
digits_std = StandardScaler().fit_transform(digits.data)
```

```python
# y데이터 = 원 핫 인코딩
from tensorflow.keras.utils import to_categorical
Y = to_categorical(digits.target)
Y[:5]
```

 - 다중 분류법

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(digits_std, Y, stratify=Y, test_size = 0.2, random_state=seed)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential([Dense(128, input_dim=64, activation='relu'),
                    Dense(64, activation='relu'),
                    Dense(10,activation='softmax')])
model.summary()
```

```python
# 다중 분류시 손실 함수 = categorical_crossentropy
model.compile('adam', 'categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
mc = ModelCheckpoint('digits_best.h5', monitor='val_loss', verbose=1, save_best_only=True)
es = EarlyStopping(patience = 30)
```

```python
hist = model.fit(X_train, Y_train, validation_split=0.2, epochs=1000, batch_size=100, verbose=0, callbacks= [mc,es])
```

```python
model.evaluate(X_test, Y_test)
```

```python
model.save('digits_best.h5')
```

```python
from tensorflow.keras.models import load_model
best_model = load_model('digits_best.h5')
best_model.evaluate(X_test, Y_test)
```
