```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
digits = load_digits()
digits.data.shape
```

```python
# x데이터 표준화
from sklearn.preprocessing import StandardScaler
digits_std = StandardScaler().fit_transform(digits.data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits_std, digits.target, test_size=0.2, random_state=seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential([Dense(100,input_dim=64, activation='relu'),
                    Dense(6, activation='relu'),
                    Dense(1)]) # 회귀 모델에서는 출력층에서 활성화 함수를 쓰지 않는다.

model.summary()
```

```python
model.compile(loss='mean_squared_error') # 회귀에서는 metric 사용하지 않음
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
mc = ModelCheckpoint('digits_best.h5', monitor='val_loss', verbose=1, save_best_only=True)
es = EarlyStopping(patience = 30)
```

```python
hist = model.fit(X_train, y_train, validation_split=0.2, epochs=1000, batch_size=100, verbose=0, callbacks= [mc,es])
```

```python
model.evaluate(X_test, y_test)
```

```python
model.save('digits_best.h5')
```

```python
from tensorflow.keras.models import load_model
best_model = load_model('digits_best.h5')
best_model.evaluate(X_test, y_test)
```
