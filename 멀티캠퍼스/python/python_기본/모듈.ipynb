- 사용자가 만든 모듈(사용자함수.py) 사용하기
  * 사용자 함수 파일 확장자: .py
  * 코드 실행 파일과 같은 폴더, 또는 하위 폴더
  * 개인 컴퓨터 (같은 폴더), 코랩은 하위 폴더
  * 코랩은 작업경로가 실행 파일 위치까지 cd로 이동하는 것이 좋다

```python
import sys
import csv

def openfile(fname):
    f = open(fname, "r")
    data = list(csv.reader(f))
    f.close()

    return data

if __name__ == "__main__":
  print(__name__)
  rdf = openfile("/content/drive/MyDrive/Multi/관광정보.csv")
  cnt = 0
  for line in rdf:
      cnt += 1
      print(line)
      if cnt == 10:
        break
```
