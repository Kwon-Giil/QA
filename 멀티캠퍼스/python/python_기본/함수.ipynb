## 함수
- 함수는 일정한 작업을 수행하는 코드 블럭
- 반복적으로 계속 사용되는 코드들을 함수로 정의하여 사용

기본 구조
def 함수명(입력 파라미터):
    문장 1
    문장 2
    [return 리턴 값]


```python
# 함수 예제
def add(a,b): # a와 b라는 2개의 파라미터(인수)를 가지는 함수
  tot = a + b
  return tot # 결과를 되돌려주는 함수

total = add(10,20)
print(total)
```

```python
# 함수 생성(정의)
def fn(i,mylist):
  i += 1 # i = i+1
  mylist.append(0)

#함수를 활용하기 위해 변수 정의 및 호출

k=10
lst=[1,2,5]

fn(k,lst) # fn 함수의 i에 k값, mylist에 lst값이 대응
print(f"k={k}\nlst= {lst}")
```

- 입력 파라미터에 기본값 지정

```python
def cal(a,b,c=1): #c 인수에 아무것도 정의하지 않았을 때 1을 초기값으로 설정
  return a*b*c

result = cal(3,4) #c 인수에 입력하지 않아도 1이 들어간 것으로 간주하고 함수 실행됨
print(result)
```

- named parameter
* 함수 호출 시 "파라미터명 = 파라미터값"과 같이
정의

```python
def report (name, age, score):
  print (f"이름:{name}, 나이:{age}, 점수: {score}")

report("윤종영",30,85) # 이렇게 쓰면 인수 순서에 대응되는 위치로만 들어감
report(name="윤종영", age = 30, score = 85) #각 인수에 맞는 위치로 찾아 들어감
```

- 가변길이 파라미터 (0~n개 만큼 닫을 수 있다)
- 파라미터 앞에 *파라미터명

```python
def plus (*nums): # 파라미터 앞에 *파라미터(인수)하면 가변길이 파라미터로 만들 수 있다
  tot = 0
  for n in nums:
    tot += n
  return tot

tot= plus(1,2)
print(tot)

t=plus(1,2,3,5,7,10,9)
print(t)
```
