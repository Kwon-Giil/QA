````python
import matplotlib.pyplot as plt #matplotlib import 해 옴

x=[0,1,2,3,4,5]
y1=[4,7,6,8,2,9]
y2=[6,9,12,18,22,14]
y3=[9,22,3,16,10,12]

plt.plot(x,y1, "r--",x,y2,"bs",x,y3,"g^") # 한 번에 여러개 출력도 가능

plt.show() # 화면에 출력
````

````python
import matplotlib.pyplot as plt #matplotlib import 해 옴

x=[0,1,2,3,4,5]
y1=[4,7,6,8,2,9]
y2=[6,9,12,18,22,14]
y3=[9,22,3,16,10,12]

plt.plot(x,y1, "r--",x,y2,"-.",x,y3,"g:") # 한 번에 여러개 출력도 가능

plt.show() # 화면에 출력
````

````python
import matplotlib.pyplot as plt #matplotlib import 해 옴

x=[0,1,2,3,4,5]
y1=[4,7,6,8,2,9]
y2=[6,9,12,18,22,14]
y3=[9,22,3,16,10,12]

plt.plot(x,y3,'r',label='y_data') #x,y3 관계와 색깔, 레이블까지 정의
plt.plot(x,y1,'g',marker='^',label='y1_data')

plt.legend() #범례 표시
plt.title('Chart Title',size=14) # 차트 제목 ('제목',글자 크기)
plt.show() # 화면에 출력
````
