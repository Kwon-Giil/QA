### 사용자 함수
- 프로그래밍에서 반복적으로 사용하는 코드를 함수로 사용자가 직접 정의
- 참조:https://wikidocs.net/24


```python
#실행문만 있는 함수
def Hello(name,age): # 함수 정의
  print(f'{name}님의 {age}세 입니다.') #f 포맷 형식을 사용하여 출력 형식 지정

Hello("홍길동",14) # 함수 호출
```

```python
## 두 수를 입력받아 더하기를 실행하고 결과를 돌려주는

def add(a, b):   # a변수, b변수 => 지역변수
    tot = a + b
    return tot

## 함수 호출
ptot=add(10, 20)
print(f"함수호출결과: {ptot}")
# print(f"함수a:{a}, 함수b:{b}") #에러



### 함수에서 여러개의 결과값 전달 #######
def add1(a, b):   # a변수, b변수 => 지역변수
    tot = a + b
    return a, b, tot

pa, pb, ptot = add1(10, 20)

print(f"함수호출결과: {ptot}")
print(f"함수a:{pa}, 함수b:{pb}")
```

```python
# 관광정보 csv 데이터 파일로 지도 그리기

# 관광정보 csv 파일 읽어오기
f = open("/content/drive/MyDrive/Multi/관광정보.csv","r")
data = f.readlines() #전체를 리스트 변수 형식으로 읽어오기 (행단위로)
f.close()
print(data[1]) # 관광정보.csv 데이터에서 위로부터 10개만 출력
data_list = data[1].replace("\n","").split(",") # 바로 위에 나눈거를 새로운 리스트에 저장
print(data_list[1].replace(",",""))
```

```python
import folium

map = folium.Map(location=[36.481074034171336, 127.28936217577971],zoom_start=10)

def map_prn(lat,lng, pop): #지도 위도, 경도 좌표와 축제 정보를 입력받은 것을 출력하는 함수
  folium.Marker([lat, lng], # 표식이 출력되는 위치
              popup = pop,                 # 마우스를 클릭하면 출력되는 내용
              icon=folium.Icon(color="red")).add_to(map)

map_prn(float(data_list[-2]),float(data_list[-1]),data_list[1])
map.save("/content/drive/MyDrive/Multi/관광맵.html")
```

```python
# 관광정보 상위 10개 데이터를 지도에 표시

import folium

f = open("/content/drive/MyDrive/Multi/관광정보.csv","r")
data = f.readlines() #전체를 리스트 변수 형식으로 읽어오기 (행단위로)
f.close()

data_lst=[data[i].replace("\n","").split(",") for i in range(1, 11)]
# print(data_lst)

map = folium.Map(location=[36.4804704, 127.289222], zoom_start=10) #맵 생성

for t in data_lst:
    try:
        if len(t) == 8:
            map_prn(float(t[-2]), float(t[-1]), t[1])
    except:
        pass

map.save("/content/drive/MyDrive/Multi/관광맵.html")
```
