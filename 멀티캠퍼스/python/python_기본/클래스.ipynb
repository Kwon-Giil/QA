### 클래스?
- 파이선은 객체지향 프로그래밍 지원
- 클래스 구조
    * class 클래스명:
        - cnt = 0 #속성

        - def __init__(self, a, b)초기자
          - self: 인스턴스 변수
          - 필요한 작업
        
        - def 함수명(): 매서드
          - 시행문
          - return 값

```python
## 클래스 연습
class myclass():
    cnt=0    # 클래스 변수

    # 초기지
    def __init__(self, 가로, 세로):
        self.width = 가로
        self.height = 세로
        myclass.cnt += 1

    # 인스턴스 메서스
    def calarea(self):
        area = self.width * self.height
        return area

    # 정적 메서드
    @staticmethod
    def isSquare(recwidth,recheight):
      return recwidth == recheight

    # 클래스 메서드
    @classmethod
    def printcnt(cls):
      print(cls.cnt)
```

```python
cls = myclass(10, 20)
print(cls.cnt)
print(cls.width)
print(cls.height)
```

```python
square = myclass.isSquare(10, 20)
print(square)
```

```python
cls1=myclass(10, 10)
cls2=myclass(10, 20)

cls1.printcnt()
cls2.printcnt()

area=cls1.calarea()
print(area)
```

```python
print(cls1.height)
cls1.height=25 # cls1의 높이를 10에서 25로 변경

print(cls1.height)
```

```python
cls2=myclass(20,30)
print(myclass.cnt)
print(cls1.cnt)
print(cls2.cnt)
```

### 실습
 - 클래스 생성
  * 객체 생성 시 이름/나이/좋아하는 것 입력받기
  * 입력 받은 정보를 이용해 화면에 출력

```python
# 클래스 생성
class Student():
    def __init__(self, name, age, like):
        self.name = name
        self.age = age
        self.like = like

    def stuinfo(self):
        print(f"이름:{self.name}, 나이:{self.age}, 좋아하는것:{self.like}")
```

```python
홍길동 = Student("홍길동", 35, "축구")
김다은 = Student("김다은", 55, "여행")

홍길동.stuinfo()
김다은.stuinfo()
```

```python
class Student_up(Student): #기존의 학생 클래스 참조해서 새로운 클래스(자식 객체) 생성
  def stuinfo(self):
    super().stuinfo()
    print(f"{self.name}, 행복한 하루 되세요.")

홍길동 = Student("홍길동", 35, "축구") #부모 클래스(Student)에 의해 만든거
김다은 = Student_up("김다은", 55, "여행") #Student에 의해 만들어진 새로운 자식 클래스 (Student_up)

홍길동.stuinfo()
김다은.stuinfo()
```
