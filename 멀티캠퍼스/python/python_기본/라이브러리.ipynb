```python
folder_name = "data_mk"
fnames=os.listdir(os.getcwd()+"/data/")
cnt=0

while True:

  if folder_name in fnames: # 해당 폴더 이름이 존재하는지 확인
    cnt +=1 # 같은 폴더 이름이 존재하면 1증가시키고
    folder_name = folder_name[:7]+str(cnt) #폴더 이름에 cnt 증가시킨거만큼 문자로 붙여라
  else:
    os.mkdir(os.getcwd()+"/data/"+folder_name)
    
    break # 반복문을 강제 종료 (없으면 무한루프 걸릴 수 있음)

print(os.listdir(os.getcwd()+"/data/"))
```

```python
# images 폴더에서 파일 가져오기
# cat 파일은 cat 폴더 생성 후 복사
# dog 파일은 dog 폴더 생성 후 복사
# 폴더가 없을 경우 생성하고, 있을 경우 해당 폴더에 저장
# 파일 저장 시 동일한 이름이 있는 경우, 뒤에 번호를 붙여라
```

```python
a = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
total = int
avg = int

def top_tier(num): # 90점 이상인 값을 골라내는 함수를 임의로 만든 거임
  if num >= 90:
    return num # 90 이상인 숫자의 값이면 출력하고
  else:
    return # 아니면 버려라

result = filter(top_tier,a) # 탑티어 함수의 결과를 저장 (필터 함수를 활용하여 탑 티어 함수의 조건을 a 리스트에 적용)
b= list(result)

total=sum(b)
avg=total/len(b)

print(avg)

```
