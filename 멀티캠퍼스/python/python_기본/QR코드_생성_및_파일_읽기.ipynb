```python
#파일 읽고 쓰기 (지금은 연결만 할거다)
#파일 객체 = open("경로/파일명","파일 열기 모드")

#qr_사이트.txt. 파일 읽어오기

# 연결된 파일을 읽어오기
# read() : 하나의 문자열로 읽어 와라 --> 하나의 문자열 형태로 묶여서 옴
# readline() : 한 줄 단위로 읽어 와라 --> 데이터의 한 줄 단위로 가져옴 (전체 가져오려면 반복적으로 수행해야 함)
# readlines() : 전체를 줄 단위로 읽어 와라 --> 전체 데이터를 리스트 단위로 물고 들어온다

f=open("/content/drive/MyDrive/Multi/qr_site.txt","r")

txt1 = f.readlines()
print(txt1)

f.close() # 열고 읽어 들어왔으면 해당 파일은 무조건 닫아야 한다
#encoding = 'UTF-8' (국제표준) / encoding='CP949' (MS 엑셀 등에서 쓰는 거)
```

```python
#qr_site.txt 파일을 읽어와서 각 사이트 별로 qr 코드 생성하기
import qrcode

f = open("/content/drive/MyDrive/Multi/qr_site.txt", "r", encoding="utf-8")
qr_txt = f.readlines()

for url in qr_txt:
    new_url = url.replace("\n", "")
    qr_img = qrcode.make(new_url)

    qr_img.save("/content/drive/MyDrive/Multi/qrimg/qr_" + new_url.split(".")[1] + ".png")

f.close()
print("작업종료")
```
