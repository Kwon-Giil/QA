```python
# 리스트 형 데이터에서 고유값 찾기
# 주어진 데이터에서 고유값을 찾아 리스트형으로 출력
# 고유 값 저장용 리스트 생성 --> for 문으로 데이터 순차적으로 가져오기
# 가져온 데이터가 생성한 리스트에 있는지 확인
# 만약 고유값 리스트 변수의 길이가 0이면 추가 아니면 확인 후 없으면 추가


indata=["사과", "배", "사과", "바나나", "귤", "바나나", "사과", "배", "바나나", "메론"]

prnlst=[]  # 빈 리스트 변수 생성

for fn in indata:
    if len(prnlst)==0: #리스트의 길이가 0이면 
        prnlst.append(fn) # 데이터 추가
    else:
        cnt=0   # 입력한 데이터의 갯수를 확인하기 윈한 변수
        for i in range(len(prnlst)):
            if fn == prnlst[i]: #리스트에 있는 데이터와 같은 지 확인
                cnt = cnt + 1  # cnt += 1 (같으면 카운트를 1 올린다.)
        if cnt == 0: # 리스트에 없는 데이터면 카운트가 안 올라가니까 리스트에 추가
            prnlst.append(fn)

print(prnlst)     

```

```python
indata=["사과", "배", "사과", "바나나", "귤", "바나나", "사과", "배", "바나나", "메론"]

prnlst=[]  # 빈 리스트 변수 생성

for fn in indata:
  if fn not in prnlst: #prnlst 변수에 없으면 추가해라
    prnlst.append(fn)

print(prnlst) 
```

```python
prnlst=list(set(indata)) #set() 함수: 집합 형식으로 제공
print(prnlst)
```

```python
text1="여러분이 그 여정의 주인공입니다. 여러분이 가는 길이 대학의 역사가 되고 여러분이 개척하는 길이 대한민국의 길이 될 것입니다."
cnt = 0 # 개수 저장할 변수 생성 (초기값으로 0)

for t in text1:
  if t == "길":
    cnt +=1 # cnt = cnt +1 

print(cnt)
```

```python
txt2=input('문자를 입력하세요')
cnt=0

for t in range(len(text1)):
  if txt2 == text1[t:t+len(txt2)]: # txt2에 입력한 문자가 text1의 t번째와 txt2의 문자 길이에 t를 더한 만큼 자른 구간에 같은 문자가 있으면
    cnt += 1

print(cnt)
```

```python
lst = [] # 단어 저장 변수
tmp = " " # 구분 단어 (여기서는 공백으로 했다)
wrd = " " # 문자를 쌓아놓는 공간

for i in text1:
  if tmp != i:
    wrd += i
  else:
    lst.append(wrd) # 쌓아놓은 문자에 " "이 들어가면 lst 변수에 저장한다 
    wrd=" "
if len(wrd) != 0:
  lst.append(wrd)

print(lst)
```
