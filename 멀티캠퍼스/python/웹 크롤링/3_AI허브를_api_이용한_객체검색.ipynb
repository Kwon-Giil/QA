 - 객체 검출 API를 이용한 이미지 사람/사물 레이블 처리

```python
import urllib3
import json
import base64
import sys
```

```python
openApiURL = "http://aiopen.etri.re.kr:8000/ObjectDetect"
accessKey = "fdf16766-d73d-48be-a4b3-c03276a6cfb2"
imageFilePath = "street1.jpg"
type = imageFilePath.split('.')[-1] # 이미지 파일의 타입
```

```python
# 이미지 파일 읽어와서 디코딩 작업
file = open(imageFilePath, "rb") # 바이너리 형식으로 만든 이미지를 갖고 들어온 거
imageContents = base64.b64encode(file.read()).decode("utf8") # 컴퓨터가 읽을 수 있게 형식을 바꿔줌
file.close()
```

```python
file = "Light betrayed me"
decoded = base64.b64encode(file.encode('ascii')).decode('ascii')
decoded
```

```python
txt = 'TGlnaHQgYmV0cmF5ZWQgbWU='
decoded = base64.b64decode(txt.encode('ascii')).decode('ascii')
decoded
```

```python
# 딕셔너리 구조의 인증 및 요청 구조 만들기
requestJson = {
    "access_key": accessKey,
    "argument": {
        "type": type,
        "file": imageContents
    }
}
```

```python
http = urllib3.PoolManager()
response = http.request(
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8"},
    body=json.dumps(requestJson))
```

```python
if response.status != 200:
  sys.exit('데이터를 받지 못했습니다.')

json_data = json.loads(response.data)
prn_data = json_data[response.data]
```
