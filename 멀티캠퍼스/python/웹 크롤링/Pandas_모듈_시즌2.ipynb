### 서울시 유가정보 가져와서 1개 파일로 합치기
1. '/content' 폴더의 'data' 폴더 생성 후 파일 업로드
2. data 폴더에 있는 데이터 읽어온 다음에 oildf에 모두 저장
3. 데이터는 concat()을 이용해 데이터 아래에 추가
4. encoding: cp949 또는 utf8 같이 있어 에러가 발생할 수 있음
  * try: ~ except: ~로 해결

- data 폴더에 있는 파일명/하위폴더명 가져오기 : os 모듈
- for 문을 이용해 첫번째 파일부터 순차적으로 읽어와 oildf에 concat 진행
- 파일 읽을 시 encoding 에러
  try:
    utf-8 형식 읽어오기
  except:
    cp949 형식 읽어오기
- 빈 DataFrame 변수 생성
  oildf=pd.DataFrame()

```python
# 서울시 유가정보 가져와 1개로 합치기
import os
# /content/data/지역_위치별(주유소)(2).csv

fnames = os.listdir("/content/data") # data 폴더 디렉토리 정의

oildf=pd.DataFrame() # 빈 데이터 프레임 만드는 거

for fname in fnames: # for문을 이용하여 폴더 안에 있는 모든 파일 불러오기
  try:
    df0 = pd.read_csv("/content/data"+fname,header=2, encoding='utf-8')
# df0에 csv 파일을 지정된 경로에 있는 파일을 utf-8 형식으로 불러와라
  except:
    df0 = pd.read_csv("/content/data"+fname,header=2, encoding='cp949')
# df0에 csv 파일을 cp949 형식으로 불러와라
    oildf=pd.concat([oildf,df0])

del df0 # 임시 데이터프레임 삭제
oildf.reset_index(drop=True, inplace=True) # oildf의 인덱스 재설정(기존에 있던 인덱스 삭제)

display(oildf.head(3))

oildf.info()
```
