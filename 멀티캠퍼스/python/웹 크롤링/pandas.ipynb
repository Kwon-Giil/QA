```python
# 기술 통계
df2=df1[["상표", "셀프여부", "경유", "휘발유"]]
df2.describe() # 현재 있는 데이터는 모두 문자(object)로 텍스트 관련 통계만 보여줌
```

```python
## 행 기준 데이터 출력(loc[], iloc[])
# loc[]
df1.loc[0]   # index 열 기준
df1.loc[1:3]
df1.loc[1:3, ["휘발유", "경유"]]
df2=df1.loc[1:10, ["셀프여부","휘발유", "경유", "상표"]]
df2
```

```python
## iloc[]: 데이터의 행/열 인덱싱된 값을 기준으로 출력 (원래 있던 데이터의 위치값으로 계산)
df2.iloc[0]
df2.iloc[1:5]
df2.iloc[1:5, 0]
df2.iloc[1:5, 1:]
df2.iloc[1:5, [0,1,3]] # 열을 찾고 싶을 때 이름이 아니라 순번을 입력해야 한다
```

```python
# df2의 경유, 휘발유 값을 정수형 데이터로 변경
# df.dtypes: 각 필드(열)의 데이터 타입을 출력
lst=[]

for i in df2.index:
  lst.append(int(df2.loc[i,["휘발유"]])) # 휘발유 필드 값을 정수형 데이터로 변경
print(lst)
df2["휘발유"] = lst # 정수형 데이터로 변경한 df2의 휘발유 필드 값을 업데이트

df2.dtypes
```

### df3 생성 후 경유(int), 휘발유 데이터(float)
- df.astype(데이터형,inplace=True), 모든 데이터를 입력한 형식으로 변경
- df = df.astype({"열이름":int, "필드명":float})

```python
# 위 코드의 for 구문을 한 줄쓰기로 변경
df3 = df1.loc[1:10,["셀프여부","휘발유", "경유", "상표"]]

df3 = df3.astype({"경유":int, "휘발유":float})
print(df3.dtypes)
```
